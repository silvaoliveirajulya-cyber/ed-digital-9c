```python
# Instalar a biblioteca inputimeout para entrada com limite de tempo
!pip install inputimeout

import random
import time
from inputimeout import inputimeout, TimeoutOccurred
from IPython.display import clear_output

def generate_sequence(length):
    return ''.join([str(random.randint(0, 9)) for _ in range(length)])

def play_memory_game():
    while True: # Loop externo para permitir o reinício do jogo
        current_sequence_length = 5
        current_input_time_limit = 5  # segundos para a entrada
        score = 0
        max_rounds = 10

        print("\n--- Jogo da Memória Numérica ---")
        print(f"Primeira sequência com {current_sequence_length} números. Você tem {current_input_time_limit} segundos para digitar.")
        time.sleep(2) # Pequena pausa antes de começar

        lost_current_game = False # Flag para indicar se a sessão atual do jogo foi perdida

        while score < max_rounds:
            clear_output(wait=True)
            print(f"\nRodada {score + 1}/{max_rounds}")
            print(f"Sequência atual: {current_sequence_length} números")
            print(f"Tempo para digitar: {current_input_time_limit} segundos")

            sequence = generate_sequence(current_sequence_length)
            print("Memorize a sequência:")
            print(f"{sequence}")

            # Tempo para o usuário memorizar
            time.sleep(current_input_time_limit)
            clear_output(wait=True)

            print("Qual foi a sequência?")
            user_input = None
            try:
                user_input = inputimeout(prompt='Sua resposta: ', timeout=current_input_time_limit)
            except TimeoutOccurred:
                print("Tempo esgotado!")
                print(f"A sequência correta era: {sequence}")
                print("\nGame Over! O jogo será reiniciado...")
                time.sleep(3)
                lost_current_game = True
                break # Sai do loop interno da rodada

            if user_input == sequence:
                score += 1
                print(f"Correto! Sua pontuação: {score}")
                if score < max_rounds:
                    current_sequence_length += 1
                    current_input_time_limit += 1 # Aumenta o tempo para a próxima rodada
                    print(f"Próxima sequência terá {current_sequence_length} números e você terá {current_input_time_limit} segundos para digitar.")
                time.sleep(2)
            else:
                print(f"Incorreto! Sua resposta: {user_input}. A sequência correta era: {sequence}")
                print("\nGame Over! O jogo será reiniciado...")
                time.sleep(3)
                lost_current_game = True
                break # Sai do loop interno da rodada

        # Após o término do loop interno `while score < max_rounds`:
        clear_output(wait=True)
        if lost_current_game:
            continue # Reinicia o loop externo `while True` para um novo jogo
        elif score == max_rounds:
            print("\nParabéns! Você ganhou!")
            print(f"Você completou {score} rodadas!")
            print("Fim de jogo!")
            break # Sai do loop externo `while True` para finalizar o jogo completamente

# Iniciar o jogo
play_memory_game
```
